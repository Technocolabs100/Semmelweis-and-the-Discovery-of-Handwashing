```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = "./cardioActivities.csv"
data = pd.read_csv(file_path,index_col='Date')
data.head()
```

```python
data.sample(5)
```

**Data preprocessing**

```python
print("Shape : ",data.shape)
data.info()
```

```python
data.describe()
```

```python
# removing the Friend's Tagged and Route Name columns because it is not used any instance in the given dataset
data.drop(["Friend's Tagged","Route Name"],axis = 1,inplace=True)
data.sample()
```

```python
# changing the name of activity type from other to unicycling
print(data.Type.value_counts())
print("Before :",data.Type.unique()) 
# method 1 
# data[data.Type == "Other"] = "Unicycling"
# method 2
data.Type.str.replace('Other',"Unicycling")
print("After : ",data.Type.unique()) 
```

```python
# missing values in the dataset 
data.isnull().sum()
```

**handling Missing values**

```python
# Average Heart Rate (bpm)
print(data['Average Heart Rate (bpm)'].unique())
avg_hr_cycle  = data[data.Type == 'Cycling']['Average Heart Rate (bpm)'].mean()
print("Average Heart Rate while cycling : ",avg_hr_cycle) 
```

```python
df_cycle = data[data.Type == 'Cycling']
df_cycle['Average Heart Rate (bpm)'].fillna(avg_hr_cycle,inplace=True)
df_cycle
```

```python
start_date = '2015-03-01'
end_date = '2018-12-31'

runs_subset_M2015_2018 = df_run.loc[(df_run.index >= start_date) & (df_run.index <= end_date)]
runs_subset_M2015_2018
```

```python
df_run_hr_all =runs_subset_M2015_2018['Average Heart Rate (bpm)']
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

# Customized x-axis tick labels
hr_zones =np.multiply(max(df_run_hr_all) , [0.5,0.7,0.85,1] )
print(hr_zones)
zone_names = ['Moderate', 'Vigorous','Extreme']
ax.set_xticklabels(zone_names, rotation=-30, ha='left')

# Create the histogram
ax.hist(df_run_hr_all, bins=hr_zones, edgecolor='black')
ax.xaxis.set(ticks = hr_zones)
# Set labels and title
ax.set_xlabel('Training Intensity')
ax.set_ylabel('Number of Activities')
ax.set_title('Heart Rate Distribution')

# Display the plot
plt.show()
```

```python
df_walk = data[data.Type == 'Walking']
df_walk.info()
```

```python
df_run_walk_cycle = pd.concat([df_cycle,df_walk,df_run])
df_run_walk_cycle.index = pd.to_datetime(df_run_walk_cycle.index)
df_run_walk_cycle.sort_index(inplace=True)
```

```python
dist_climb_cols = ['Distance (km)', 'Climb (m)']
df_totals = df_run_walk_cycle.groupby('Type')[dist_climb_cols].sum()
df_totals
```

```python
df_run_walk_cycle.pivot_table(index = 'Type',values=dist_climb_cols,aggfunc = 'sum')
```

```python
df_run_walk_cycle.pivot_table(index = 'Type',values=dist_climb_cols,aggfunc = 'describe').stack()
```
