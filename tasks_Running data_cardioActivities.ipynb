```python
# Define the columns to drop
cols_to_drop = ['Route Name', "Friend's Tagged", 'Notes']  # Replace with actual column names
```

```python
# Delete unnecessary columns
df_activities = df_activities.drop(columns=cols_to_drop)
```

```python
# Calculate activity type counts
activity_counts = df_activities['Type'].value_counts()
print(activity_counts)
```

```python
# Rename 'Other' values to 'Unicycling'
df_activities['Type'] = df_activities['Type'].str.replace('Other', 'Unicycling')
```

```python
# Count missing values in each column
missing_values_count = df_activities.isnull().sum()
print(missing_values_count)
```

Task 3

```python
# Calculate the sample mean for Average Heart Rate (bpm) for 'Cycling' activity type
avg_hr_cycle = df_activities[df_activities['Type'] == 'Cycling']['Average Heart Rate (bpm)'].mean()
avg_hr_cycle
```

```python
# Filter df_activities for 'Cycling' activity type
df_cycle = df_activities[df_activities['Type'] == 'Cycling'].copy()
```

```python
# Fill missing values for Average Heart Rate (bpm) in df_cycle
df_cycle['Average Heart Rate (bpm)'].fillna(int(avg_hr_cycle), inplace=True)
```

```python
# Count missing values for all columns in df_cycle
missing_values_count = df_cycle.isnull().sum()
print(missing_values_count)
```

Task 6

```python
import matplotlib.pyplot as plt

# Select distance and heart rate information
runs_distance = runs_subset_2015_2018['Distance (km)']
runs_hr = runs_subset_2015_2018['Average Heart Rate (bpm)']

# Create subplots with shared x-axis
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(12, 8))

# Plot distance on the first subplot
ax1.plot(runs_distance)
ax1.set_ylabel('Distance (km)')
ax1.set_title('Distance and Heart Rate')

# Plot heart rate on the second subplot
ax2.plot(runs_hr, color='red')
ax2.axhline(runs_hr.mean(), color='blue', linewidth=1, linestyle='-.')
ax2.set_ylabel('Heart Rate (bpm)')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Subset df_activities for data from 2013 through 2018 and select Distance (km) column
df_run_dist_2013_2018 = df_activities.loc['2013':'2018', 'Distance (km)']

# Count annual totals with resample() and sum()
df_run_dist_annual = df_run_dist_2013_2018.resample('A').sum()

# Create a plot with specified size
plt.figure(figsize=(8.0, 5.0))

# Customize the plot with horizontal span
ax = plt.gca()
ax.axhspan(0, 800, color='red', alpha=0.2)

# Plot the annual totals
df_run_dist_annual.plot(kind='bar', ax=ax)
plt.xlabel('Year')
plt.ylabel('Total Distance (km)')
plt.title('Annual Total Distance')

# Show the plot
plt.show()
```
